```python
import os
from PIL import Image
```

```python
old_path = '/data/file/image13_2label/train/'
new_path = '/data/file/image13_2label/train_img/'

img = '/data/file/image13_2label/train/2.jpg'
new_name = '/data/file/image13_2label/train_img/2.jpg'
```

```python
def move_img(old_path, new_path):
    f_list = os.listdir(old_path)
    for i in f_list:
        # os.path.splitext: 分离文件名与拓展名
        if os.path.splitext(i)[1] == '.jpg':
            old_img = old_path+i
            new_img = new_path+i
            os.rename(old_img, new_img)
            print(new_img)
```

```python
move_img(old_path, new_path)
```
